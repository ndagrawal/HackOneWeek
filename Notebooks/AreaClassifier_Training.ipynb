```python
%%time
import numpy as np

fields = [
    'System.Id',
    'System.Title',
    'System.AreaPath',
    'System.AreaId',
    'System.TeamProject',
    'System.WorkItemType', 
    'System.Tags',
    'System.Description',
    'Microsoft.VSTS.TCM.ReproSteps',
    'Microsoft.DevDiv.WalkThroughDescription'
]

# make sure this something like 100*k + 1
start_id = 1
# make sure this is always a multiple of 100
end_id = 1500000

# default batch size
batch_size = 200

all_wits_dictionary = {}

while start_id<=end_id:
    ids = range(start_id, start_id+batch_size)
        
    fetched_wits = wit_client.get_work_items(ids=ids, fields=fields, error_policy='omit')
    
    #store work items in a dictionary
    for wit in fetched_wits:
        if wit.fields['System.WorkItemType'] in ['Bug', 'User Story', 'Scenario', 'Feature', 'DTS Task', 'Feedback Ticket']:
            all_wits_dictionary[wit.id] = wit
    
    print('fetched ' + str(len(fetched_wits)) + ' work items with ids in range: '+ str(np.min(ids)) + ' - '+ str(np.max(ids)))
    start_id += batch_size

    
total = str(len(all_wits_dictionary.keys()))
print(total + ' work items fetched')
```

```python
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
import re


URLS = r"""                             # Capture 1: entire matched URL
  (?:
  https?:                               # URL protocol and colon
    (?:
      /{1,3}                            # 1-3 slashes
      |                                 #   or
      [a-z0-9%]                         # Single letter or digit or '%'
                                        # (Trying not to match e.g. "URI::Escape")
    )
    |                                   #   or
                                        # looks like domain name followed by a slash:
    [a-z0-9.\-]+[.]
    (?:[a-z]{2,13})
    /
  )
  (?:                                   # One or more:
    [^\s()<>{}\[\]]+                    # Run of non-space, non-()<>{}[]
    |                                   #   or
    \([^\s()]*?\([^\s()]+\)[^\s()]*?\)  # balanced parens, one level deep: (...(...)...)
    |
    \([^\s]+?\)                         # balanced parens, non-recursive: (...)
  )+
  (?:                                   # End with:
    \([^\s()]*?\([^\s()]+\)[^\s()]*?\)  # balanced parens, one level deep: (...(...)...)
    |
    \([^\s]+?\)                         # balanced parens, non-recursive: (...)
    |                                   #   or
    [^\s`!()\[\]{};:'".,<>?]      # not a space or one of these punct chars
  )
  |                                     # OR, the following to match naked domains:
  (?:
          (?<!@)                        # not preceded by a @, avoid matching foo@_gmail.com_
    [a-z0-9]+
    (?:[.\-][a-z0-9]+)*
    [.]
    (?:[a-z]{2,13})
    \b
    /?
    (?!@)                               # not succeeded by a @,
                                        # avoid matching "foo.na" in "foo.na@example.com"
  )
"""

URL_RE = re.compile(URLS, re.VERBOSE | re.I | re.UNICODE)
HANG_RE = re.compile(r'([^a-zA-Z0-9])\1{3,}')


def too_many_digits(text):
    return len([c for c in text if str.isdigit(c)])>0

def camel_case_handler(name):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower().replace('_', ' ')


def to_tokens(text, stem=True):
    text = str(text)
    text = BeautifulSoup(text, 'lxml').get_text(' ')
    text = re.sub('<[^>\s]+>', ' ', text)
    text = BeautifulSoup(text, 'lxml').text
    text = URL_RE.sub(' ', text)
    text = re.sub('[\w.+-]+@[\w-]+\.(?:[\w-]\.?)+[\w-]', '', text)
    text = re.sub('(?:(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.){3}(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)', '', text)
    text = re.sub('<[^>]*>', ' ', text)
    text = HANG_RE.sub(r'\1\1\1', text)
    text = re.sub('[\W_]+', ' ', text)
    text = text.lower()
    words = text.split()
    text = ' '.join([camel_case_handler(t) for t in words if not str.isdigit(t) and not too_many_digits(t)])
    words = text.split()
    stemmer = SnowballStemmer('english')
    stops = set(stopwords.words('english'))
    words = [w for w in words if not w in stops and len(w) > 1]

    if stem:
        words = [stemmer.stem(word) for word in words]

    return words
```

```python
%%time
import pandas as pd


text_fields = [
    'System.Title',
    'System.Description',
    'Microsoft.VSTS.TCM.ReproSteps',
    'Microsoft.DevDiv.WalkThroughDescription',
    'System.Tags'
]

def get_text(wid):
    wit = all_wits_dictionary[wid]
    fieldtext = ' '.join([wit.fields[f] if f in wit.fields else '' for f in text_fields])
    tokens = to_tokens(fieldtext)
    return ' '.join(tokens), len(tokens)

wit_ids = [id for id in all_wits_dictionary]
wit_texts = [get_text(id) for id in all_wits_dictionary]
wit_areas = [all_wits_dictionary[id].fields['System.AreaPath'] for id in all_wits_dictionary]
wit_types = [all_wits_dictionary[id].fields['System.WorkItemType'] for id in all_wits_dictionary]
d = {'id': wit_ids, 'area': wit_areas, 'text': [tuple[0] for tuple in wit_texts], 'token_count': [tuple[1] for tuple in wit_texts], 'wit_type': wit_types }
df = pd.DataFrame(d)
```

```python
def area_1(area):
  return area.split('\\')[0]

df2 = df
df2 = df2.drop(df2[df2.token_count<2].index)
df2 = df2.drop(df2[df2.area.str.contains('eprecated')==True].index)
```

```python
df3.groupby('area').count()
def shorten_area(area):
  return '\\'.join(area.split('\\')[0:4])

df3['short_area'] = df3['area'].apply(shorten_area)
```

```python

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
X_text = df4['text']
y_area = df4['short_area']
```

```python
le.fit(y_area)
y_area_encoded = le.transform(y_area)
```

```python
le.classes_
```

```python
%%time

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_text, y_area_encoded, test_size=0.20, random_state=90)
```

```python
%%time

from sklearn import metrics
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python

nb_clf = Pipeline([('tfidf', TfidfVectorizer(max_features=2000,use_idf=True,smooth_idf=True,ngram_range=(1,2))), ('clf', MultinomialNB(alpha=1e-2))])
```

```python
%%time
nb_clf.fit(X_train, y_train)
predicted = nb_clf.predict(X_test)
print(metrics.classification_report(y_test, predicted, target_names=le.classes_))
```
